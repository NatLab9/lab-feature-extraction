```python
#2. Plot all the categorical variables with the proper plot. What can you see?
def plot_discrete_variables(categorical):
  
    for col in categorical.columns:
        plt.figure()
        plt.title(col)
        categorical[col].value_counts().plot(kind='bar')

plot_discrete_variables(categorical)
```

```python
#4. Plot time variable. Can you extract something from it?

# Convert the 'effective_to_date' column to datetime
categorical['effective_to_date'] = pd.to_datetime(categorical['effective_to_date'])

# Create a new column 'month' with the month values
categorical['month'] = categorical['effective_to_date'].dt.month

# Display the DataFrame with the new 'month' column
categorical.head()
```

```python
# Create a new column 'month' by applying the 'dt.month' attribute to the 'effective_to_date' column
categorical['month'] = categorical['effective_to_date'].apply(lambda x: x.month)
```
